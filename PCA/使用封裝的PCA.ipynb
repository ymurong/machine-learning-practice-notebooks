## 从高维数据向低维数据的映射


```python
import matplotlib.pyplot as plt
import numpy as np

X = np.empty((100, 2))
np.random.seed(666)
X[:, 0] = np.random.uniform(0., 100., size=100)
X[:, 1] = 0.75 * X[:, 0] + 3. + np.random.normal(0., 10., size=100)


def demean(X):
    return X - np.mean(X, axis=0)


X_demean = demean(X)
plt.scatter(X_demean[:, 0], X_demean[:, 1])
```

```python
from lib.PCA import PCA

pca = PCA(n_components=2)
pca.fit(X)
```

```python
pca.components_
```

```python
plt.scatter(X_demean[:, 0], X_demean[:, 1])
for i in range(pca.components_.shape[0]):
    plt.plot([0, pca.components_[i, 0] * 30], [0, pca.components_[i, 1] * 30])
```

## 使用隨機梯度下降

```python
from lib.PCA import PCA

pca = PCA(n_components=2)
pca.fit(X, n_iters=5, type="sga")
```

```python
pca.components_
```

```python
plt.scatter(X_demean[:, 0], X_demean[:, 1])
for i in range(pca.components_.shape[0]):
    plt.plot([0, pca.components_[i, 0] * 30], [0, pca.components_[i, 1] * 30])
```
